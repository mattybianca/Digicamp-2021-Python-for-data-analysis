````python
import pandas as pd
````

````python
df = pd.read_csv('Farm_production2015_2020.csv')
````

#### A short description of the dataset:
This is a dataset that contains the daily production of a large farm in the period between January 2015 and December 2020. It produces Beer, Eggs, Wheat, and Milk. 

````python
df
````

````python
df['Product'].unique()
````

````python
df.columns
````

Why doesn't ```columns```  have the parenthesis at the end?? Well, as the function ```.index```, it refers to an index object so in this case we don't use the parenthesis

````python
df.index
````

Now we are going to create a new dataframe that contains only the Eggs production

Copy is a function that allows you to modify a subset without changing the starting dataset.
Furthermore this practice avoids you warnings from pandas. Do you know the difference between warning and error?
<img style="float: left;" src="copydf.png" width=400 height=300>

````python
df_eggs = df.copy()[df['Product'] == 'Eggs']
````

````python
df_eggs
````

````python
df_eggs.set_index('Date',inplace = True)
````

Why I didn't set the dates as index previously? Because there were multiple element with the same date.

````python
df_eggs.index
````

````python
df_eggs
````

````python
df_eggs.drop('Product', axis = 1, inplace = True)
````

````python
df_eggs
````

````python
df_eggs['SMA_14'] = df_eggs['Production_quantity'].rolling(14).mean()
````

````python
df_eggs
````

````python
df_eggs.dropna(inplace = True)
````

````python
df_eggs
````

````python
round(df_eggs['SMA_14'],0)
````

````python
df_eggs
````

````python
df_eggs['SMA_14'] = round(df_eggs['SMA_14'],0)
````

````python
df_eggs
````

````python
df_eggs['Production_quantity'].std()
````

````python
round(df_eggs['Production_quantity'].std())
````
